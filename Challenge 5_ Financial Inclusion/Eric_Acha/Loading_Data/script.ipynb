```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.colors import ListedColormap
import seaborn as sns

#LOADING DATA
df = pd.read_csv("D://Download//farmers_texts.csv") 
```

```python
#Remove Irrelevant Columns
df = df[['question_language',
       'question_content', 'question_sent', 'response_language', 'response_content',
       'response_sent','question_user_created_at', 'question_user_country_code',
       'response_user_country_code', 'response_user_created_at']]
df.shape
```

```python
#Proportion of the different questions based on language
df['question_language'].value_counts()
```

```python
#Stratified Random Sampling
eng = 12 #Proportions for each language
swa = 6
nyn = 2
lug = 1 

sampleMultiplier = 150000 #multiplier to get a large number of samples while preserving proportion

dfEng = df[df['question_language'] == 'eng'].sample(n=eng*sampleMultiplier, random_state=42) #strata
dfSwa = df[df['question_language'] == 'swa'].sample(n=swa*sampleMultiplier, random_state=42)
dfNyn = df[df['question_language'] == 'nyn'].sample(n=nyn*sampleMultiplier, random_state=42)
dfLug = df[df['question_language'] == 'lug'].sample(n=lug*sampleMultiplier, random_state=42)

textSample = pd.concat([dfEng, dfSwa, dfNyn, dfLug], ignore_index=True)
textSample.to_csv('textSample.csv', index=False)
```

```python
#Count Alpha Numeric Characters in The Text (this is for translation)
tempText = " ".join(textSample['question_content'])

charCount = sum(1 for char in tempText if char.isalnum())
print("Number of Characters: " + str(charCount))
```
