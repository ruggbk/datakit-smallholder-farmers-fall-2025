```python
import duckdb
```

```python
duckdb.query("""
COPY (
    SELECT *
    FROM read_csv_auto(
        'C:/Brandon/farmersdirect_data_b0cd514b-b9cc-4972-a0c2-c91726e6d825.csv',
        ignore_errors = true,
        sample_size = -1,
        all_varchar = true
    )
) TO 'C:/Brandon/farmersdirect_data_b0cd514b-b9cc-4972-a0c2-c91726e6d825_v2.parquet' 
(FORMAT PARQUET, OVERWRITE TRUE);
""")


```

```python
# Count rows in original CSV
original_rows = duckdb.query("""
    SELECT COUNT(*) FROM read_csv_auto(
        'C:/Brandon/farmersdirect_data_b0cd514b-b9cc-4972-a0c2-c91726e6d825.csv',
        ignore_errors = true,
        sample_size = -1
    )
""").fetchall()[0][0]
```

```python
# Count rows in Parquet
parquet_rows = duckdb.query("""
    SELECT COUNT(*) FROM 'C:/Brandon/farmersdirect_data_b0cd514b-b9cc-4972-a0c2-c91726e6d825_v2.parquet'
""").fetchall()[0][0]

print(f"Original rows (with ignored errors): {original_rows}")
print(f"Parquet rows: {parquet_rows}")
```

```python
unique_question_count = duckdb.query("""
    SELECT COUNT(DISTINCT question_content)
    FROM 'C:/Brandon/farmersdirect_data_b0cd514b-b9cc-4972-a0c2-c91726e6d825_v2.parquet'
""").fetchall()[0][0]
print(unique_question_count)
```

```python
unique_rows_maincolumns = duckdb.query("""
    SELECT COUNT(DISTINCT ROW(question_id, question_content, question_topic,response_id, response_content))
    FROM 'C:/Brandon/farmersdirect_data_b0cd514b-b9cc-4972-a0c2-c91726e6d825_v2.parquet'
""").fetchall()[0][0]
print(unique_rows_maincolumns)
```

```python
unique_response_count = duckdb.query("""
    SELECT COUNT(DISTINCT response_id)
    FROM 'C:/Brandon/farmersdirect_data_b0cd514b-b9cc-4972-a0c2-c91726e6d825_v2.parquet'
""").fetchall()[0][0]
print('# of Unique Responses: ' + str(unique_response_count))
```
